```python
import pandas as pd
import sweetviz as sv
```

```python
df1 = pd.read_csv("../input/titanic/train.csv")
report1 = sv.analyze(df1,target_feat='Survived')
report1.show_html('Report1.html')
```

```python
df2 = pd.read_csv("../input/titanic/test.csv")
report2 = sv.compare([df1, "Train"], [df2, "Test"])
report2.show_html('Report2.html')
```

```python
df3 = pd.read_csv("../input/tabular-playground-series-oct-2022/train_0.csv")
report3 = sv.analyze(df3)
report3.show_html('Report3.html')
```
